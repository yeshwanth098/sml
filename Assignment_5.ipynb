```python
import pandas as pd
df = pd.read_csv('/content/diabetes_data_upload.csv')
df
```

```python
m=df['class']
print(m)
```

```python
n=df.drop('class',axis=1)
print(n)
```

```python
import pandas as pd
for col in n.columns:
  if n[col].dtype == 'object':
try:
      n[col] = pd.to_numeric(n[col], errors='coerce')
    except:
        print(f"Could not convert column {col} to numeric. It might contain non-numerical values.")
        n = n.fillna(0)
        pp=(n-n.min())/(n.max()-n.min())
        print(pp)
```

```python
 import matplotlib.pyplot as plt
 plt.plot(n,m)
 plt.xlabel('x- axis')
 plt.ylabel('y- axis')
 plt.title('Diabetis')
 plt.show()
```
