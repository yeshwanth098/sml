```python
import pandas as pd
d = pd.read_csv("/content/train.csv")
print(d)
```

```python
y=d['price_range']
print(y)
```

```python
x=d.drop('price_range',axis=1)
print(x)
```

```python
dd=(x-x.min())/(x.max()-x.min())
print(dd)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dd,y,test_size=0.2,random_state=20)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
